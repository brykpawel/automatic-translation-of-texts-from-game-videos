```python
from fairseq.models import BaseFairseqModel

model_path = "/content/drive/My Drive/english-polish-conv"
model = BaseFairseqModel.from_pretrained(
    model_name_or_path=model_path,
    checkpoint_file="checkpoint_best.pt",
    data_name_or_path=model_path,
    tokenizer="moses",
    bpe="subword_nmt",
    bpe_codes="code",
    cpu=True
)

# Pass the sentence as a list
import time
start_time = time.time()
translated_sentence = model.translate(sentences=["A team of astronomers discovered an extraordinary planet in the constellation of Virgo."], beam=5)
print("--- %s seconds ---" % (time.time() - start_time))
print(translated_sentence)
```

```python
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
def calculate_bleu(groundtrouth, recognition):

    # Tokenize the reference and hypothesis texts
    recognition = recognition.lower().split()
    groundtrouth = groundtrouth.lower().split()

    # Calculate BLEU score
    smoothing_function = SmoothingFunction().method1
    bleu_score = sentence_bleu([groundtrouth], recognition, smoothing_function=smoothing_function)

    return bleu_score
```

```python
import pandas as pd
import time
# Wczytanie pliku TSV
file_path = '/content/drive/My Drive/Pary zdań w angielski-polski - 2024-06-19.tsv'
df = pd.read_csv(file_path, sep='\t', header=None)
bleu_scores = 0
i = 0
start = time.time()
# Iteracja przez wiersze i przypisanie wartości do zmiennych
for index, row in df.iterrows():
    if i<1000:
      english = row[1]
      polish = row[3]
      prediction = model.translate(sentences=english, beam=5)

      # print(prediction)
      bleu = calculate_bleu(polish, prediction)
      print(bleu, i)
      bleu_scores += bleu
      i+=1
    else:
      break
    # print(calculate_bleu(polish, polish))
    # print(f'English: {english}, Polish: {polish}')
print("bleu_scores:", bleu_scores/i)
print("time:", time.time()-start)
```
